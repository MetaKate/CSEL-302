# **1. Import Libraries**

```python
import pandas as pd #Data Manipulation
import numpy as np #Numerical Operation
import matplotlib.pyplot as plt #Data Visualization
from sklearn.linear_model import LinearRegression #Data Modeling
```

# **2. General Random Data:**

```python
num_points = 50

X = np.random.randint(0, 100, 50)
Y = 2 * X + np.random.normal(0,10,num_points)
```

# **3. Creating DataFrame and Saving to CSV**

```python
data = pd.DataFrame({'X': X,'Y':Y}) #TableData X and Header Y

data.to_csv('data.csv', index=False) #Ginagamit sa taas para walang space

print(data.head())
```

# **4. Visualizing the Data**

```python
plt.scatter(data['X'],data['Y'])
plt.xlabel('Independent Variable (X)')
plt.ylabel('Dependent Variable (Y)')
plt.title('Scatter Plot of data')
plt.show()
```

# **5. Preparing Data for Modeling**

```python
X = data[['X']] #Independent Variable
Y = data[['Y']] #Dependent Variable
```

# **6. Fitting the Linear Regression Model**

```python
model = LinearRegression()

model.fit(X, Y)
```

# **7. Getting Model Coefficients**

```python
slope = model.coef_[0]

intercept = model.intercept_

print('Slope', slope)
print('Intercept', intercept)
```

# **8. Visualizing the Liner Regression Line**

```python
plt.scatter(X,Y)
plt.plot(X,model.predict(X), color='red')

plt.xlabel('Independet Variable')
plt.ylabel('Dependent Variable')
plt.title('Linear Regression of Data')
plt.show()
```

# **9. Making Prediction**

```python
new_X = [[np.random.randint(100,200)]]

predicted_Y = model.predict(new_X)

print('Value of X: ' ,new_X)
print('Predicted Value of Y: ' ,predicted_Y)
```
