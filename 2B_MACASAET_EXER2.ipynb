```python
import pandas as pd
import matplotlib.pyplot as plt
```

1. Data Loading

```python
data = pd.read_csv('hardwareStore.csv')
data.head()
```

2. Initial Data Analysis

```python
hardware = data.groupby('CATEGORY_NAME')['QUANTITY'].sum()

standard_cost_mean = data['STANDARD_COST'].mean()
standard_cost_median = data['STANDARD_COST'].median()
standard_cost_std = data['STANDARD_COST'].std()
print("STANDARD COST")
print("MEAN = " , standard_cost_mean)
print("MEDIAN = " , standard_cost_median)
print("SD = " , standard_cost_std)
print(" ")
list_price_mean = data['LIST_PRICE'].mean()
list_price_median = data['LIST_PRICE'].median()
list_price_std = data['LIST_PRICE'].std()
print("LIST PRICE")
print("MEAN = " , list_price_mean)
print("MEDIAN = " , list_price_median)
print("SD = " , list_price_std)
print(" ")
quantity_mean = data['QUANTITY'].mean()
quantity_median = data['QUANTITY'].median()
quantity_std = data['QUANTITY'].std()
print("QUANTITY")
print("MEAN = " , quantity_mean)
print("MEDIAN = " , quantity_median)
print("SD = " , quantity_std)

print(" ")
uni_categories = data.groupby('PRODUCT_ID')['PRODUCT_ID'].count()
print(uni_categories)
```

3. Category Analysis

```python
hardware.plot(kind='bar', color='lightpink')
plt.xlabel('')
plt.ylabel('Number of Products')
plt.title('COMPUTER PARTS', fontweight='bold', color='black')
plt.xticks(rotation = 0)
plt.show
```
