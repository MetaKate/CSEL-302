Import Libraries

```python
import pandas as pd #Data Manipulation
import numpy as np #Numerical Operation
import matplotlib.pyplot as plt #Data Visualization
from sklearn.linear_model import LinearRegression #Data Modeling
```

Data Analysis

```python
data = pd.read_csv('Lesson_4_student_grades_dataset.csv')
data.head()
```

Data Visualizing

```python
plt.scatter(data['Previous_Exam_Scores'], data['Grades'])
plt.xlabel('Independet Variable')
plt.ylabel('Dependent Variable')
plt.title('Linear Regression of Data')
plt.show()
```

Preparing Data for Modeling

```python
X = data[['Previous_Exam_Scores']] #Independent Variable
Y = data[['Grades']] #Dependent Variable
```

Fitting the Linear Regression Model

```python
model = LinearRegression()

model.fit(X, Y)
```

Getting Model Coefficients

```python
slope = model.coef_[0]

intercept = model.intercept_

print('Slope', slope)
print('Intercept', intercept)
```

```python
plt.scatter(X,Y)
plt.plot(X,model.predict(X), color='red')

plt.xlabel('Independet Variable')
plt.ylabel('Dependent Variable')
plt.title('Linear Regression of Data')
plt.show()
```

```python
new_X = [[np.random.randint(0,100)]]

predicted_Y = model.predict(new_X)

print('Value of X: ' ,new_X)
print('Predicted Value of Y: ' ,predicted_Y)
```
